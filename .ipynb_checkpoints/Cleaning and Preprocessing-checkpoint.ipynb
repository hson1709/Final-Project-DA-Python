```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv(r"D:\Xử lý dữ liệu với Python\Raw Data Theanh28.Hanoi.csv")
```

## Drop tất cả các cột chứa toàn bộ giá trị NaN

```python
nan_df = df.isna()
all_nan_columns = nan_df.all()
columns_with_all_nan = all_nan_columns[all_nan_columns].index.tolist()
print("Các cột có tất cả giá trị là NaN:", columns_with_all_nan)
```

```python
columns_with_all_nan = df.columns[df.isnull().all()]
df = df.drop(columns=columns_with_all_nan)
df
```

## Loại bỏ các hàng và các cột dữ liệu bị trùng lặp

```python
df = df.drop_duplicates()
df
```

```python
df = df.T.drop_duplicates().T
df
```

## Loại bỏ hastag

```python
print(df.at[2,'text'])
print(df.at[31,'text'])
```

```python
def remove_text_and_hashtags(text):
    # Loại bỏ đoạn text sau dấu (*)
    text = re.sub(r'\s*\(\*\).*$', '', text)
    # Loại bỏ các hashtag
    text = re.sub(r'#\w+', '', text)
    return text
df['text'] = df['text'].apply(remove_text_and_hashtags)

print(df.at[2,'text'])
print(df.at[31,'text'])
```

## Loại bỏ các emoji

```python
emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F"  
                            u"\U0001F300-\U0001F5FF"  
                            u"\U0001F680-\U0001F6FF"  
                            u"\U0001F1E0-\U0001F1FF"  
                            u"\U00002500-\U00002BEF"  
                            u"\U00002702-\U000027B0"
                            u"\U00002702-\U000027B0"
                            u"\U000024C2-\U0001F251"
                            u"\U0001f926-\U0001f937"
                            u"\U00010000-\U0010ffff"
                            u"\u2640-\u2642"
                            u"\u2600-\u2B55"
                            u"\u200d"
                            u"\u23cf"
                            u"\u23e9"
                            u"\u231a"
                            u"\ufe0f"  
                            u"\u3030"
                            u"\u23F0"
                            "]+", flags=re.UNICODE)

df['text'] = df['text'].apply(lambda x: emoji_pattern.sub(r'', x))
print(df.at[2,'text'])
print(df.at[31,'text'])
```

```python
print("Đoạn text 1:", end="\n")
print(df.at[18,'text'])
print("Đoạn text 2:", end="\n")
print(df.at[31,'text'])
```

## Loại bỏ từ khóa "Xem thêm"

```python
def remove_xem_them(text):
    return text.replace("Xem thêm", "")
df['text'] = df['text'].apply(remove_xem_them)
print("Đoạn text 1:", end="\n")
print(df.at[18,'text'])
print("Đoạn text 2:", end="\n")
print(df.at[31,'text'])
```

## Loại bỏ những trường thông tin không cần thiết hoặc không có giá trị phân tích

```python
# Danh sách tên cột cần xóa
columns_to_drop = [
    'image', 'image_lowquality', 'images', 'images_lowquality',
    'video', 'video_id', 'video_thumbnail', 'links', 'image_id', 'image_ids',
    'user_id', 'username', 'user_url', 'page_id', 'w3_fb_url',
    'images_lowquality_description', 'images_description',
    'is_live', 'available', 'fetched_time'
]
df = df.drop(columns=columns_to_drop)
```

```python
print(df.columns)
```

## Xử lý trường dữ liệu “reactor” để tách riêng các loại reaction:

```python
df1 = df
df1['reactions'] = df1['reactions'].apply(lambda x : dict(eval(x)))
reactions = df1['reactions'].apply(pd.Series)
```

```python
# tạo dataframe mới 
cleaned_data = pd.concat([df1, reactions], axis=1).drop('reactions', axis=1)
cleaned_data[['text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                             'shares','comments','reaction_count']]
```

```python
cleaned_data[['text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                             'shares','comments','reaction_count']] = cleaned_data[['text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                             'shares','comments','reaction_count']].fillna(0)

cleaned_data[['text','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                             'shares','comments','reaction_count']]
```

## Chuyển dữ liệu cột time sang datatime

```python
cleaned_data['time'] = pd.to_datetime(cleaned_data['time'])
print(cleaned_data['time'].dtype)
```

## Xuất file data sau khi đã xử lý:

```python
csv_file_path = r"D:\Xử lý dữ liệu với Python\New Data Theanh28.Hanoi.csv"
cleaned_data.to_csv(csv_file_path, index=False)
```
