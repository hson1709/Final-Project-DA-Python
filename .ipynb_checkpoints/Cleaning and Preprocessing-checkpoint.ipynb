```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv(r"D:\X·ª≠ l√Ω d·ªØ li·ªáu v·ªõi Python\Raw Data Theanh28.Hanoi.csv")
```

## Drop t·∫•t c·∫£ c√°c c·ªôt ch·ª©a to√†n b·ªô gi√° tr·ªã NaN

```python
nan_df = df.isna()
all_nan_columns = nan_df.all()
columns_with_all_nan = all_nan_columns[all_nan_columns].index.tolist()
print("C√°c c·ªôt c√≥ t·∫•t c·∫£ gi√° tr·ªã l√† NaN:", columns_with_all_nan)
```

```python
columns_with_all_nan = df.columns[df.isnull().all()]
df = df.drop(columns=columns_with_all_nan)
df
```

## Lo·∫°i b·ªè c√°c h√†ng v√† c√°c c·ªôt d·ªØ li·ªáu b·ªã tr√πng l·∫∑p

```python
df = df.drop_duplicates()
df
```

```python
df = df.T.drop_duplicates().T
df
```

## Lo·∫°i b·ªè hastag

```python
print(df.at[2,'text'])
print(df.at[31,'text'])
```

```python
def remove_text_and_hashtags(text):
    # Lo·∫°i b·ªè ƒëo·∫°n text sau d·∫•u (*)
    text = re.sub(r'\s*\(\*\).*$', '', text)
    # Lo·∫°i b·ªè c√°c hashtag
    text = re.sub(r'#\w+', '', text)
    return text
df['text'] = df['text'].apply(remove_text_and_hashtags)

print(df.at[2,'text'])
print(df.at[31,'text'])
```

## Lo·∫°i b·ªè c√°c emoji

```python
emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F"  
                            u"\U0001F300-\U0001F5FF"  
                            u"\U0001F680-\U0001F6FF"  
                            u"\U0001F1E0-\U0001F1FF"  
                            u"\U00002500-\U00002BEF"  
                            u"\U00002702-\U000027B0"
                            u"\U00002702-\U000027B0"
                            u"\U000024C2-\U0001F251"
                            u"\U0001f926-\U0001f937"
                            u"\U00010000-\U0010ffff"
                            u"\u2640-\u2642"
                            u"\u2600-\u2B55"
                            u"\u200d"
                            u"\u23cf"
                            u"\u23e9"
                            u"\u231a"
                            u"\ufe0f"  
                            u"\u3030"
                            u"\u23F0"
                            "]+", flags=re.UNICODE)

df['text'] = df['text'].apply(lambda x: emoji_pattern.sub(r'', x))
print(df.at[2,'text'])
print(df.at[31,'text'])
```

```python
print("ƒêo·∫°n text 1:", end="\n")
print(df.at[18,'text'])
print("ƒêo·∫°n text 2:", end="\n")
print(df.at[31,'text'])
```

## Lo·∫°i b·ªè t·ª´ kh√≥a "Xem th√™m"

```python
def remove_xem_them(text):
    return text.replace("Xem th√™m", "")
df['text'] = df['text'].apply(remove_xem_them)
print("ƒêo·∫°n text 1:", end="\n")
print(df.at[18,'text'])
print("ƒêo·∫°n text 2:", end="\n")
print(df.at[31,'text'])
```

## Lo·∫°i b·ªè nh·ªØng tr∆∞·ªùng th√¥ng tin kh√¥ng c·∫ßn thi·∫øt ho·∫∑c kh√¥ng c√≥ gi√° tr·ªã ph√¢n t√≠ch

```python
# Danh s√°ch t√™n c·ªôt c·∫ßn x√≥a
columns_to_drop = [
    'image', 'image_lowquality', 'images', 'images_lowquality',
    'video', 'video_id', 'video_thumbnail', 'links', 'image_id', 'image_ids',
    'user_id', 'username', 'user_url', 'page_id', 'w3_fb_url',
    'images_lowquality_description', 'images_description',
    'is_live', 'available', 'fetched_time'
]
df = df.drop(columns=columns_to_drop)
```

```python
print(df.columns)
```

## X·ª≠ l√Ω tr∆∞·ªùng d·ªØ li·ªáu ‚Äúreactor‚Äù ƒë·ªÉ t√°ch ri√™ng c√°c lo·∫°i reaction:

```python
df1 = df
df1['reactions'] = df1['reactions'].apply(lambda x : dict(eval(x)))
reactions = df1['reactions'].apply(pd.Series)
```

```python
# t·∫°o dataframe m·ªõi 
cleaned_data = pd.concat([df1, reactions], axis=1).drop('reactions', axis=1)
cleaned_data[['text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                             'shares','comments','reaction_count']]
```

```python
cleaned_data[['text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                             'shares','comments','reaction_count']] = cleaned_data[['text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                             'shares','comments','reaction_count']].fillna(0)

cleaned_data[['text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                             'shares','comments','reaction_count']]
```
